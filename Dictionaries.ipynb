Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates. **bold text**

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

```python
#Dictionary Items
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict["brand"])
print(thisdict["year"])
```

```python
#length of dict
print(len(thisdict))
```

```python
#Get Keys
thisdict.keys()
```

```python
#check item in dict
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
if "model" in thisdict:
  print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

```python
if "age" in thisdict:
  print("yes present in dict")
else:
  print("not present")
```

```python
#change the value of dictionary
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018
```

```python
print(thisdict["year"])
```

Loop Through a Dictionary

```python
for item in thisdict:
  print(item)
```

```python
for x in thisdict:
  print(thisdict[x])
```

```python
for x in thisdict.items():
  print(x)
```

```python
for x ,y in thisdict.items():
  print(x,y)
```

Functions


A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

```python
def my_function():
  print("Simple Function")

my_function()
```

Arguments

Information can be passed into functions as arguments.

Arguments are specified after the function name, inside the parentheses

```python
def my_function(name):
  print(name )

my_function("ali")
my_function("amin")
my_function("jon")
```

```python
def my_function(food):
  for x in food:
    print(x)

fruits = ["apple", "banana", "cherry"]

my_function(fruits)
```
